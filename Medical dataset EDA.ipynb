<h1 style="color:green"><b>Importing Modules</b> </h1>

## Import libraries

```python
# Numpy
import numpy as np

# Pandas
import pandas as pd

# MatPlotLib
import matplotlib.pyplot as plt

# Seaborn
import seaborn as sns

# scipy
from scipy import stats
```

## Import data

```python
df = pd.read_csv(r"C:\Users\gurte\OneDrive\All About Learning\Coding Dojo\Learning Material\02 Weekly Assignments\wk16 03-06 (Enrichment, wk3, hypothesis)\medicaldataset-eda\medical_data.csv")
df.columns= df.columns.str.strip().str.lower()
df=df[['vitd_levels', 'doc_visits', 'totalcharge']]
df
```

```python
df.info()
```

<h1 style="color:green"><b>EDA: Vitd Levels</b> </h1>

## plot

```python
# column name 
col1 = 'vitd_levels'
feature1 = df[col1]

# calcualtions 
mean = feature1.mean()
median = feature1.median()
std = feature1.std()
plus_one_std = mean + std
minus_one_std = mean - std

# plot 
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(data=df, x=feature1, kde=False, stat='density')
sns.kdeplot(data=df, x=feature1, ax=ax, color='black')
ax.axvline(mean, color = 'red', label=f'Mean = {mean:,.2f}')
ax.axvline(median, color = 'green', ls='--', label = f'Median = {median:,.2f}')
ax.axvline(plus_one_std, color = 'black', label = f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)

ax.set_title(f'{col1}')
ax.legend();
```

```python
kurt = stats.kurtosis(feature1, fisher = False)
kurt
```

## interpretation

This is a continious feature and has a normal distribution. Kurtosis is more than 3 so it is leptokurtic.  

<h1 style="color:green"><b>EDA: Doc Visits</b> </h1>

## plot

```python
# column name 
col2 = 'doc_visits'
feature2 = df[col2]

# calcualtions 
mean = feature2.mean()
median = feature2.median()
std = feature2.std()
plus_one_std = mean + std
minus_one_std = mean - std

# plot 
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(data=df, x=feature2, discrete=True, kde=False, stat='density')
sns.kdeplot(data=df, x=feature2, ax=ax, color='black')
ax.axvline(mean, color = 'red', label=f'Mean = {mean:,.2f}')
ax.axvline(median, color = 'green', ls='--', label = f'Median = {median:,.2f}')
ax.axvline(plus_one_std, color = 'black', label = f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)

ax.set_title(f'{col2}')
ax.legend();
```

```python
kurt2 = stats.kurtosis(feature2, fisher = False)
kurt2
```

## interpretation

This is a discrete feature and has a normal distribution. Kurtosis is ~3 so it is Mesokurtic.
